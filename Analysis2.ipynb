```python
# %% [markdown]
# # Overwatch Hero Psychology: Archetype Analysis
# **Modules Applied:**
# 1. Similarity (Cosine Similarity)
# 2. Dimensionality Reduction (PCA)
# 3. Clustering (K-Means)
# 4. Visualization (Radar Charts)

# %% [code]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

# Machine Learning Imports
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Styling
plt.rcParams["figure.figsize"] = (10, 8)
sns.set_style("whitegrid")
# Custom Overwatch-ish palette (Tank=Blue, Damage=Red, Support=Green)
role_colors = {"Tank": "#1f77b4", "Damage": "#d62728", "Support": "#2ca02c"}

# %% [markdown]
# ## 1. Load & Preprocess Data
# We load the dataset and normalize the psychology stats (Mobility, Sustain, Mech Skill, HP).

# %% [code]
# Load Data
df = pd.read_csv("overwatch_hero_psychology_full.csv")
df.head()

# Create a 'Role' column for coloring plots (reverse the one-hot encoding)
def get_role(row):
    if row['Tank'] == 1: return 'Tank'
    if row['Damage'] == 1: return 'Damage'
    if row['Support'] == 1: return 'Support'
    return 'Unknown'

df['Role'] = df.apply(get_role, axis=1)

# Features to analyze
features = ["Mobility", "Self_Sustain", "Mechanical_Skill", "HP"]
X = df[features]

# Scaling (StandardScaler is crucial for PCA and Clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a DataFrame for scaled data to make plotting easier later
df_scaled = pd.DataFrame(X_scaled, columns=features)
df_scaled['Hero'] = df['Hero']
df_scaled['Role'] = df['Role']

print(f"Loaded {len(df)} heroes.")
df.head()

# %% [markdown]
# ## 2. Similarity Analysis (Cosine)
# Who is psychologically similar to whom?

# %% [code]
# Compute Cosine Similarity
similarity_matrix = cosine_similarity(X_scaled)
sim_df = pd.DataFrame(similarity_matrix, index=df['Hero'], columns=df['Hero'])

# Visualize: Heatmap of Similarity
plt.figure(figsize=(16, 14))
sns.heatmap(sim_df, cmap='coolwarm', center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title("Hero Psychological Similarity Matrix", fontsize=16)
plt.show()

# Quick check: Top 3 most similar heroes to 'Hazard' (if he exists in your list)
target_hero = "Hazard"
if target_hero in sim_df.index:
    print(f"Top matches for {target_hero}:")
    print(sim_df[target_hero].sort_values(ascending=False).iloc[1:4])

# %% [markdown]
# ## 3. Dimensionality Reduction (PCA)
# Compressing our 4D stats into a 2D map of the 'Hero Galaxy'.

# %% [code]
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# Add PCA coordinates to our dataframe
df['PC1'] = principal_components[:, 0]
df['PC2'] = principal_components[:, 1]

# Plot PCA
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Role', data=df, palette=role_colors, s=100, alpha=0.8)

# Add labels to points
for i, txt in enumerate(df['Hero']):
    plt.annotate(txt, (df.PC1[i]+0.05, df.PC2[i]+0.05), fontsize=9)

plt.title('Overwatch Hero Space (PCA)', fontsize=16)
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)', fontsize=12)
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)', fontsize=12)
plt.legend(title='Role')
plt.show()

# %% [markdown]
# ## 4. Clustering (K-Means)
# Finding the latent "Archetypes" regardless of official Role.
# *Are there "Tanky DPS" or "Evasive Supports"?*

# %% [code]
# Determine optimal clusters (Elbow Method could be used here, but we'll choose 4 for archetypes)
k = 4
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Plot Clusters on PCA Map
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', style='Role', data=df, palette='viridis', s=120)

# Add labels
for i, txt in enumerate(df['Hero']):
    plt.annotate(txt, (df.PC1[i], df.PC2[i]), fontsize=8, alpha=0.7)

plt.title(f'K-Means Clustering (k={k}) on Hero Psychology', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()

# Show who is in each cluster
for i in range(k):
    print(f"\n--- Cluster {i} ---")
    print(df[df['Cluster'] == i]['Hero'].values)

# %% [markdown]
# ## 5. Visualizing Archetypes (Radar Charts)
# **"Can we do better?"** Yes. Radar charts visualize the *personality* of each cluster.

# %% [code]
# 1. Calculate the mean values for each cluster (on original 0-10 scale for readability)
cluster_means = df.groupby('Cluster')[features].mean().reset_index()

# 2. Function to create Radar Chart
def create_radar_chart(row, categories, title, ax, color):
    # Number of variables
    N = len(categories)

    # Values for this cluster (repeat first value to close the loop)
    values = row[categories].values.flatten().tolist()
    values += values[:1]

    # Angles for each axis
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Plot
    ax.plot(angles, values, linewidth=2, linestyle='solid', color=color, label=title)
    ax.fill(angles, values, color=color, alpha=0.25)

    # Fix axis labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, size=10)
    ax.set_ylim(0, 10)  # Assuming max rating is 10 (HP might need norm if not 1-10 scaled)
    ax.set_title(title, size=14, color=color, y=1.1)

# Normalize HP for visual consistency in Radar (0-10 scale approximation)
# We divide HP by 100 just for the chart so it fits with 1-10 stats
radar_df = cluster_means.copy()
radar_df['HP'] = radar_df['HP'] / 75  # Rough scaling: 750HP -> 10

# Create Subplots
fig, axes = plt.subplots(1, k, subplot_kw=dict(polar=True), figsize=(20, 5))
colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF'] # Custom colors

feature_labels = ["Mobility", "Sustain", "Mech. Skill", "HP"]

for i in range(k):
    create_radar_chart(
        radar_df.iloc[i],
        features,
        f"Archetype {i}",
        axes[i],
        colors[i]
    )

plt.tight_layout()
plt.show()
```
